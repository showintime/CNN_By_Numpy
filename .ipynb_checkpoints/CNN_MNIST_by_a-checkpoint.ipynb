```python

import numpy

import math

import struct

import random

import time

import cv2
import pickle
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
```

```python
'''
image like (28x28x3)  HWC
filters like (28x28x3x50)  HWCN
stride is ...
number is equal to filters.shape[-1]  N
'''

#这里可以考虑先把所有的receptive field整合为一个更大更大的矩阵，然后把Filters也整合，成为一个大的矩阵   然后点击运算
def convolution(image,filters,filtersb,stride):
    #print(image.shape)
    image=image.reshape((image.shape[0],image.shape[1],image.shape[2],1))
    
    feature_maps=numpy.empty(((image.shape[0]-filters.shape[0])//stride[0]+1,(image.shape[1]-filters.shape[1])//stride[1]+1,filters.shape[-1]))
    
    for i in range(0,image.shape[0]-filters.shape[0]+1,stride[0]):
        for j in range(0,image.shape[1]-filters.shape[1]+1,stride[1]):
            
            receptive_field=image[i:i+filters.shape[0],j:j+filters.shape[1]]
            
            tem=receptive_field*filters
            
            feature_maps[i//stride[0],j//stride[1]]=tem.sum(axis=(0,1,2))+filtersb
            
    return feature_maps
            
```

```python
def pooling(image,shape,stride):
    pooling_maps=numpy.empty(((image.shape[0]-shape[0])//stride[0]+1,(image.shape[1]-shape[1])//stride[1]+1,image.shape[2]))
    
    for i in range(0,image.shape[0]-shape[0]+1,stride[0]):
        for j in range(0,image.shape[1]-shape[1]+1,stride[1]):
            
            pooling_field=image[i:i+shape[0],j:j+shape[1]]
            
            pooling_maps[i//stride[0],j//stride[1]]=numpy.amax(pooling_field,axis=(0,1))
            
    return pooling_maps
```

```python
#Full Connextion Layer
#FC (Inner Product)
def InnerProduct(image,w,b):
    sum_axis=tuple(range(len(image.shape)))
    
    tem_shape=list(image.shape) 
    tem_shape.append(1)
    image=image.reshape(tuple(tem_shape))
    
    tem=image*w
    
    return tem.sum(axis=sum_axis)+b
```

```python
def ReLU(image):
    return (numpy.abs(image)+image)/2
```

```python
def Softmax(image):
    tem=numpy.exp(image)
    return tem/tem.sum()
```

```python
def Loss(y,label):
    dif=y-label
    tem=dif*dif
    return tem.sum()
```

```python
def dLoss(y,label):
    dif=y-label
    return 2*dif
```

```python
def dReLU(dL,image):
    tem=1*(image>0)
    return dL*tem
```

```python
def dSoftmax(dL,image):
    tem=Softmax(image)
    dS=numpy.diag(tem)-numpy.outer(tem,tem)
    #这里先用矩阵乘法，以后可能要改成点积和
    return numpy.dot(dL,dS)
```

```python
def dInnerProduct(dL,image,w,b):
    
    tem=dL*w
        
    tem_shape=list(image.shape)
    tem_shape.append(1)
    image=image.reshape(tuple(tem_shape))
    
    
    #这里默认全连接层的后一层输出的是一个维度的数组
    #return dL/dimage, dL/dw, dL/db
    return tem.sum(axis=(len(w.shape)-1)),image*dL,dL
```

```python
def dpooling(dL,image,shape,stride):
    #如果这里shape的大小  小于 等于  stride的大小，我们就可以不用初值为零   然后相加
    dp=numpy.zeros_like(image)
    
    for i in range(0,image.shape[0]-shape[0]+1,stride[0]):
        for j in range(0,image.shape[1]-shape[1]+1,stride[1]):
            
            pooling_field=image[i:i+shape[0],j:j+shape[1]]
            
            #先判断相等，也就是只有最大值的时候为True，其他情况都是False，然后乘1，转成数字数组（True=1，False=0）
            
            dpooling_field=(pooling_field==numpy.amax(pooling_field,axis=(0,1)))*dL[i//stride[0],j//stride[1]]
            dp[i:i+shape[0],j:j+shape[1]]+=dpooling_field
            
    return dp
```

```python
def dconvolution(dL,image,filters,filtersb,stride,isfirst):
    
    dimage=numpy.zeros_like(image)
    dfilters=numpy.zeros_like(filters)
    
    image=image.reshape((image.shape[0],image.shape[1],image.shape[2],1))
    for i in range(0,image.shape[0]-filters.shape[0]+1,stride[0]):
        for j in range(0,image.shape[1]-filters.shape[1]+1,stride[1]):
            
            tem=dL[i//stride[0],j//stride[1]]*filters
            
            if not isfirst:
                dimage[i:i+filters.shape[0],j:j+filters.shape[1]]+=tem.sum(axis=(len(filters.shape)-1))
            
            receptive_field=image[i:i+filters.shape[0],j:j+filters.shape[1]]
            dfilters+=dL[i//stride[0],j//stride[1]]*receptive_field
            
            
    
    
    return dimage,dfilters,dL.sum(axis=(0,1))
    
```

```python
para={}
dist=[-math.sqrt(6/(784+10)),math.sqrt(6/(784+10))]

para['filters1']=numpy.random.rand(5,5,1,20)*(dist[1]-dist[0])+dist[0]
para['filters1b']=numpy.zeros(20)
para['filters1stride']=(1,1)

para['pooling1shape']=(2,2)
para['pooling1stride']=(2,2)

para['filters2']=numpy.random.rand(5,5,20,50)*(dist[1]-dist[0])+dist[0]
para['filters2b']=numpy.zeros(50)
para['filters2stride']=(1,1)

para['pooling2shape']=(2,2)
para['pooling2stride']=(2,2)

para['w1']=numpy.random.rand(4,4,50,500)*(dist[1]-dist[0])+dist[0]
para['b1']=numpy.zeros(500)

para['w2']=numpy.random.rand(500,10)*(dist[1]-dist[0])+dist[0]
para['b2']=numpy.zeros(10)
```

```python
caltime={}
caltime['c1']=0
caltime['c2']=0
caltime['p2']=0
caltime['p1']=0
caltime['In1']=0
caltime['In2']=0
caltime['Re']=0
caltime['Softmax']=0
caltime['dc1']=0
caltime['dc2']=0
caltime['dp2']=0
caltime['dp1']=0
caltime['dIn1']=0
caltime['dIn2']=0
caltime['dRe']=0
caltime['dSoftmax']=0
caltime['dLoss']=0
```

```python
def calculate_a_grad(image,label):
    
    #print('in calc',image.shape)

    time0=time.time()
    
    image=image.reshape(28,28,1)
    dpara={}
    c1=convolution(image,para['filters1'],para['filters1b'],para['filters1stride'])
    
    time1=time.time()
    caltime['c1']+=time1-time0
    
    p1=pooling(c1,para['pooling1shape'],para['pooling1stride'])
    
    time0=time.time()
    caltime['p1']+=time0-time1
    
    c2=convolution(p1,para['filters2'],para['filters2b'],para['filters2stride'])
    
    time1=time.time()
    caltime['c2']+=time1-time0
    
    p2=pooling(c2,para['pooling2shape'],para['pooling2stride'])
    
    time0=time.time()
    caltime['p2']+=time0-time1
    
    inner1=InnerProduct(p2,para['w1'],para['b1'])
    
    time1=time.time()
    caltime['In1']+=time1-time0
    
    #print('relu::',p1)
    relu1=ReLU(inner1)
    
    time0=time.time()
    caltime['Re']+=time0-time1
    
    inner2=InnerProduct(relu1,para['w2'],para['b2'])
    
    time1=time.time()
    caltime['In2']+=time1-time0
    
    y=Softmax(inner2)
    
    
    time0=time.time()
    caltime['Softmax']+=time0-time1
    
    dL=dLoss(y,label)
    time1=time.time()
    caltime['dLoss']+=time1-time0
    
    dL=dSoftmax(dL,inner2)
    
    time0=time.time()
    caltime['dSoftmax']+=time0-time1
    
    dL,dpara['w2'],dpara['b2']=dInnerProduct(dL,relu1,para['w2'],para['b2'])
    
    time1=time.time()
    caltime['dIn2']+=time1-time0
    
    dL=dReLU(dL,inner1)
    time0=time.time()
    caltime['dRe']+=time0-time1
    
    dL,dpara['w1'],dpara['b1']=dInnerProduct(dL,p2,para['w1'],para['b1'])
    time1=time.time()
    caltime['dIn1']+=time1-time0
    
    dL=dpooling(dL,c2,para['pooling2shape'],para['pooling2stride'])
    
    time0=time.time()
    caltime['dp2']+=time0-time1
    
    dL,dpara['filters2'],dpara['filters2b']=dconvolution(dL,p1,para['filters2'],para['filters2b'],para['filters2stride'],False)
    
    time1=time.time()
    caltime['dc2']+=time1-time0
    
    dL=dpooling(dL,c1,para['pooling1shape'],para['pooling1stride'])
    
    time0=time.time()
    caltime['dp1']+=time0-time1
    
    dL,dpara['filters1'],dpara['filters1b']=dconvolution(dL,image,para['filters1'],para['filters1b'],para['filters1stride'],True)
    
    time1=time.time()
    caltime['dc1']+=time1-time0
    
    return dpara
    
    
```

```python
label_keys=numpy.identity(10)
def calculate_banch_grad(banch_time,banch_size):
    
    banch_grad=calculate_a_grad(train_image[banch_size*banch_time],train_label[banch_size*banch_time])
    for i in range(banch_size*banch_time+1,banch_size*(banch_time+1)):
        
        #print('训练'+str(i+1)+'/'+str(banch_size)+'个参数')
        timg=train_image[i]
        timg=timg.reshape(28,28,1)
        tem_grad=calculate_a_grad(timg,label_keys[train_label[i]])
        for j in banch_grad.keys():
            banch_grad[j]+=tem_grad[j]
    for j in banch_grad.keys():
        banch_grad[j]/=banch_size
        
    return banch_grad
```

```python
def train_banch(banch_grad,learn_rate):
    #print('正在修正参数')
    for j in banch_grad.keys():
        para[j]-=learn_rate*banch_grad[j]
    
            
```

```python
def train_epoch():
    banch_size=100
    learn_rate=0.01
    for i in tqdm_notebook(range(train_num//banch_size)):
        
        #print('#===============开始训练第'+str(i+1)+'/'+str(train_num//banch_size)+'组==============#')
        banch_grad=calculate_banch_grad(i,banch_size)
        train_banch(banch_grad,learn_rate)
```

```python
def train(epoch_num):
    for i in tqdm_notebook(range(epoch_num)):
        train_epoch()
        ver[0].append(validate_verify(para))
        ver[1].append(test_verify(para))
        ver[2].append(train_verify(para))
```

```python
def validate_verify(parameters):
    error_num=0
    for i in range(validate_num):
        
        if predict(validate_image[i]).argmax()!=validate_label[i]:
        
            error_num+=1
            
    rate=1-error_num/validate_num
    #print('验证集准确率：'+str(rate*100)+'%')
    return rate
```

```python
def test_verify(parameters):
    error_num=0
    for i in range(test_num):
        
        if predict(test_image[i]).argmax()!=test_label[i]:
        
            error_num+=1
            
    rate=1-error_num/test_num
    #print('验证集准确率：'+str(rate*100)+'%')
    return rate
```

```python
def train_verify(parameters):
    error_num=0
    for i in range(train_num):
        
        if predict(train_image[i]).argmax()!=train_label[i]:
        
            error_num+=1
            
    rate=1-error_num/train_num
    #print('验证集准确率：'+str(rate*100)+'%')
    return rate
```

```python
def predict(image):
    image=image.reshape(28,28,1)
    
    c1=convolution(image,para['filters1'],para['filters1b'],para['filters1stride'])
    p1=pooling(c1,para['pooling1shape'],para['pooling1stride'])
    c2=convolution(p1,para['filters2'],para['filters2b'],para['filters2stride'])
    p2=pooling(c2,para['pooling2shape'],para['pooling2stride'])
    inner1=InnerProduct(p2,para['w1'],para['b1'])
    relu1=inner1
    #relu1=ReLU(inner1)
    inner2=InnerProduct(relu1,para['w2'],para['b2'])
    y=Softmax(inner2)
    
    return y
```
